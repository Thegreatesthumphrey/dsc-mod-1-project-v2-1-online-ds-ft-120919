## Exploring the significance of movie genre

### Importing necessary libraries // Establish a connection to the database

```python
from importlib import reload

import sqlite3
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt

import seaborn as sns
```

```python
conn = sqlite3.connect("movies_db.sqlite")
cur = conn.cursor()
```

```python
for row in cur.execute("SELECT name FROM sqlite_master WHERE type='table';"):
    print(row)
```

### Joining necessary tables

```python
cur.execute('''SELECT *
               FROM imdb_title_basics
               JOIN tn_movie_budgets
               ON imdb_title_basics.original_title = tn_movie_budgets.movie
               JOIN tmdb_movies
               ON tn_movie_budgets.movie = tmdb_movies.original_title
            ''')
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df.head()
```

```python
df.shape
```

### Looking at the relationship between genre and worldwide gross

```python
df['genres'] = df['genres'].map(lambda x: x.split(",") if x else x)
```

```python
df.head()
```

```python
cur.execute('''SELECT *
               FROM imdb_title_basics
               JOIN tn_movie_budgets
               ON imdb_title_basics.original_title = tn_movie_budgets.movie
               JOIN tmdb_movies
               ON tn_movie_budgets.movie = tmdb_movies.original_title
               ORDER BY popularity DESC
               LIMIT 150
            ''')
df2 = pd.DataFrame(cur.fetchall())
df2.columns = [i[0] for i in cur.description]
df2.head()
```

```python
df2['genres'] = df2['genres'].map(lambda x: x.split(",") if x else x)
```

```python
df2.head()
```

```python
all_genres =[]
none_counter = 0
for genre in df2['genres']:
    if genre == None:
        none_counter += 1
    else:
        all_genres += genre
```

```python
all_genres
```

```python
unique_genres = set(all_genres)
```

```python
genre_count = {}
for genre in unique_genres:
    genre_count[genre] = all_genres.count(genre)
```

```python
genre_count
```

```python
values = []
labels = []

for genre, count in genre_count.items():
    labels.append(genre)
    values.append(count)
```

```python
plt.figure(figsize=(8, 5))
plt.barh(labels, values)
plt.title("Genre Counts\nFor Top 150\nProfitable Movies")
```

As far as worldwide gross goes, the most popular genres by far are Adventure, Action, and Drama followed closely by Sci-Fi and Fantasy.

### Looking at the relationship between genre and popularity

```python
cur.execute('''SELECT *
               FROM imdb_title_basics
               JOIN tn_movie_budgets
               ON imdb_title_basics.original_title = tn_movie_budgets.movie
               JOIN tmdb_movies
               ON tn_movie_budgets.movie = tmdb_movies.original_title
               ORDER BY worldwide_gross DESC
               LIMIT 150
            ''')
df3 = pd.DataFrame(cur.fetchall())
df3.columns = [i[0] for i in cur.description]
df3.head()
```

```python
df3['genres'] = df3['genres'].map(lambda x: x.split(",") if x else x)
df3.head()
```

```python
all_genres2 =[]
none_counter = 0
for genre in df3['genres']:
    if genre == None:
        none_counter += 1
    else:
        all_genres2 += genre
```

```python
unique_genres2 = set(all_genres2)
```

```python
genre_count2 = {}
for genre in unique_genres2:
    genre_count2[genre] = all_genres2.count(genre)
```

```python
genre_count2
```

```python
values2 = []
labels2 = []

for genre, count in genre_count2.items():
    labels2.append(genre)
    values2.append(count)
```

```python
plt.figure(figsize=(8, 5))
plt.barh(labels2, values2)
plt.title("Genre Counts\nFor Top 150\nMost Popular Movies")
```

Again, Adventure and Action seem to be the most popular with Sci-Fi and Fantasy close behind. However, unlike the most profitable movies, Comedy and Animation also rank fairly high.
