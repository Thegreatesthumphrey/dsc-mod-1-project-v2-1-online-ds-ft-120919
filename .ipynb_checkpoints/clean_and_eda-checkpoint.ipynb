# Import Necessary Libraries

```python
from importlib import reload

import sqlite3
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt

import seaborn as sns
```

# Connect to Database & Create 'Cursor Object'

```python
conn = sqlite3.connect("movies_db.sqlite")
```

```python
cur = conn.cursor()
```

# Get List of All Tables in DB

```python
for row in cur.execute("SELECT name FROM sqlite_master WHERE type='table';"):
    print(row)
```

# 'tmdb_movies' - CLEAN

```python
cur.execute('''SELECT * FROM tmdb_movies''')
df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df.head()
```

```python
print(df.isna().sum()) # We have no missing values.
print(df.dtypes) # All data types seem OK, except for the date column, going to convert to a datetime object
```

# Convert 'release_date' column from Object to datetime object

```python
df['release_date'].dtype
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])  # Convert to datetime object
```

```python
df.head() 
```

# Create 'release_day' column to indicate day of week movie was released

```python
df['release_day'] = df['release_date'].map(lambda x: x.weekday())
df.head()
```

# Keep only movies with 'original_language' equal to English

```python
df.groupby(df['original_language']).count()
```

```python
print(df.shape)
print(df.loc[df['original_language'] == 'en'].shape)
```

```python
df = df.loc[df['original_language'] == 'en']
df.head()
```

# Create 'release_month' column

```python
df['release_month'] = df['release_date'].map(lambda x: x.month)
df.head()
```

# 'tmdb_movies' - EDA

```python
plt.figure(figsize=(13, 8))
sns.boxplot(x='release_day', y='popularity', data=df)
plt.show()
```

```python
plt.figure(figsize=(13, 8))
sns.boxplot(x='release_month', y='popularity', data=df)
plt.show()
```

```python
plt.figure(figsize=(13, 8))
sns.distplot(df['vote_average'], bins = 10)
plt.show()

plt.figure(figsize=(13, 8))
sns.distplot(df['popularity'], bins = 80)
plt.show()
```

```python
print(df['vote_average'].mean())
print(df['vote_average'].mode())
print(df['vote_average'].median())
print(df['vote_average'].std())
```

# 'tn_movie_budgets' - CLEAN/EDA

```python
cur.execute('''SELECT * FROM tn_movie_budgets''')
df3 = pd.DataFrame(cur.fetchall())
df3.columns = [x[0] for x in cur.description]
df3.head()
```

```python
print(df3.isna().sum())
print(df3.shape) # No missing values
```

```python
df3.info() # I want to convert some of these into more appropriate data types
```

# Convert all money columns into integers

```python
df3['production_budget'] = df3['production_budget'].str.replace("$","").str.replace(",","").astype('int')
```

```python
df3['domestic_gross'] = df3['domestic_gross'].str.replace("$","").str.replace(",","").astype('int')
df3['worldwide_gross'] = df3['worldwide_gross'].str.replace("$","").str.replace(",","").astype('int')
df3.info()
```

# Convert date column into datetime object

```python
df3['release_date'] = pd.to_datetime(df['release_date'])
df3.info()
```

```python
df3.head()
```

# Add a few calculated columns
- domestic_net column (domestic_gross - production_budget)
- worldwide_net column (worldwide_gross - production_budget)
- release_day and release_month columns (for any analysis on performance by day/month)

```python
df3['domestic_net'] = df3['domestic_gross'] - df3['production_budget']
df3.head()
```

```python
df3['domestic_worldwide'] = df3['worldwide_gross'] - df3['production_budget']
df3['release_day'] = df3['release_date'].map(lambda x: x.weekday())
df3['release_month'] = df3['release_date'].map(lambda x: x.month)
df3.head()
```

```python
df3 = df3.rename(columns={"domestic_worldwide": "worldwide_net"})
df3.head()
```

# Explore relationship between release day/month and gross profit

```python
plt.figure(figsize=(13, 8))
sns.boxplot(x='release_day', y='worldwide_gross', data=df3)
plt.show()
```

```python
plt.figure(figsize=(13, 8))
sns.boxplot(x='release_month', y='worldwide_gross', data=df3)
plt.show()

plt.figure(figsize=(13, 8))
sns.boxplot(x='release_month', y='worldwide_net', data=df3)
plt.show()
```

# Look at correlation between budget and worldwide gross

```python
plt.figure(figsize=(13, 8))
sns.regplot(x='production_budget', y='worldwide_gross', data=df3)
plt.show()
```
